```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2020,3,1)
end = datetime.datetime(2021,10,10)
```

# Gather Stock Information

```python


# Amazon
AMZN = data.DataReader('AMZN','yahoo',start,end)
# Alibaba
BABA = data.DataReader('BABA','yahoo',start,end)
# Walmart
WMT = data.DataReader('WMT','yahoo', start,end)
# Apple
AAPL = data.DataReader('AAPL','yahoo', start,end)
# JD
JD = data.DataReader('JD','yahoo', start,end)
```

```python
tickers = ['AMZN','BABA','WMT','AAPL','JD'] 
```

```python
consumer_stocks = pd.concat([AMZN,BABA,WMT,AAPL,JD], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Consumer Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Consumer_Stock.csv')
```
